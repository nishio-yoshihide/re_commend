```python
import pandas as pd
from kedro.framework.session import KedroSession
from kedro.framework.session.session import _activate_session
from pathlib import Path
import sys

sys_path_orig = sys.path
sys.path.append("../src/")
project_path = Path("../")
session = KedroSession.create("re_commend", project_path)
_activate_session(session, force=True)
context = session.load_context()
catalog = context.catalog
```

```python
brands = catalog.load("brand_dataset")
brands
```

```python
from sklearn.manifold import TSNE
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

mat = np.vstack(brands["vector"].values)
print(mat.shape)
tsne = TSNE().fit_transform(mat)
plt.plot(tsne[:, 0], tsne[:, 1], ".")
```

```python
breed_count = brands.groupby("breed").size().reset_index().rename(columns={0:"count"})
data = (
    brands[["brand", "breed"]]
    .merge(breed_count, on="breed")
    .assign(
        x=tsne[:,0],
        y=tsne[:,1],
        breed=lambda df: [
            breed if count > 1 else "その他"
            for breed, count in zip(df["breed"], df["count"])
        ])
)

sns.set(font='Yu Gothic')
sns.scatterplot(data=data, x="x", y="y", hue="breed")
```
